## **I. Comprender la poblaci칩n**

Para visualizar las variaciones en los valores de un dataset se puede utilizar:

*   Histograma
*   Boxplot
*   Bar Chart



### **Ejemplo para Atributos Num칠ricos**

```python
numerical_attributes = ["age", "hoursPerWeek"]

for att in numerical_attributes:
  plt.subplot(2,1,1)
  df[att].plot.hist()
  plt.subplot(2,1,2)
  df[att].plot.box(vert=False)
  plt.tight_layout()
  plt.show()
```

### **Ejemplo de Comparaci칩n usando Diagramas de Barras**

**TO DO:** Crea una visualizaci칩n que usa diagrama de barras para comparar el atributo categ칩rico raza en base a sus ingresos.

**Primera manera:** Colocar un diagrama de barra despu칠s del otro

```python
df.info()
```

```python
for i,poss in enumerate(income_list):
  plt.subplot(2,1,i+1)
  BM = df.income == poss
  df[BM].race.value_counts().plot.barh()
  plt.xlim([0,25000])
  plt.ylabel(poss)

plt.show()
```

**Segunda manera:** Mezclar ambos histogramas en uno solo. Note los labels de `income,race`

```python
df.groupby(["race","income"]).size().plot.barh()

plt.show()
```

**Tercera manera:** Mezcla de dos histogramas usando leyendas con el uso de diferentes colores.

```python
df.groupby(["income","race"]).size().unstack().plot.barh()

plt.show()
```

**Cuarta manera:** Usaremos un **stacked bar chart**


```python
df.groupby(["race","income"]).size().unstack().plot.barh(stacked=True)

plt.legend(loc=4)

plt.show()
```
