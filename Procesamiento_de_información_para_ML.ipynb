```python
# Imputaci√≥n con modelo regresi√≥n lineal

data_1_df_imp_lm = data_1_df.copy()
data_1_df_imp_lm[var_[0]].describe().transpose()
```

```python
#Separamos los datos en train y test (Holdout Method)
X_train, X_test, y_train, y_test = train_test_split(data_1_df_imp_lm_1[var_[1:]], # Features
                                                    data_1_df_imp_lm_1["Appliances"], # Target
                                                    test_size=0.35, # % de datos para el testeo
                                                    random_state=123) # Colocar una semilla aleatoria
```

```python
#Considerar los siguientes objetos: X_train, X_test, y_train, y_test

#Llamar a la libreria que contiene el algoritmo
from sklearn import linear_model

#Entrenamos al algoritmo
##Definir un objetivo (imputer) y los hiperpar√°metros del modelo
modelo_rl = linear_model.LinearRegression()
#Entrenamiento del algoritmo
modelo_rl.fit(X=X_train, y = y_train)
```

```python
#El algoritmo entrenado es modelo_rl
#El algoritmo entrenado va generar predicciones sobre los datos que tiene info faltante (Appliances)
#El algoritmo entrenado necesita los features para generar las predicciones (data_1_df_imp_lm[var_[1:])
var_imp = modelo_rl.predict(data_1_df_imp_lm[var_[1:]])[data_1_df_imp_lm['Appliances'].isnull()]
```

```python

# Imputaci√≥n con modelo Random Forest Regressor

## Paso 1: Una copia del dataframe inicial
data_1_df_imp_rf = data_1_df.copy()

## Paso 2: Resumen descriptivo de la variable a imputar
print(f"Resumen de la variable antes de ser imputada: \n{data_1_df_imp_rf[var_[0]].describe().transpose()}\n")


## Paso 3: Definir los casos que seran usados para entrenar y testar el modelo (casos con informaci√≥n a nivel de Appliances)
data_1_df_imp_rf_1 = data_1_df_imp_rf[~data_1_df_imp_rf.Appliances.isna()]

## Paso 4: Separamos los datos en train y test (Holdout Method)
X_train, X_test, y_train, y_test = train_test_split(data_1_df_imp_rf_1[var_[1:]], # Features
                                                    data_1_df_imp_rf_1["Appliances"], # Target
                                                    test_size=0.35, # % de datos para el testeo
                                                    random_state=123) # Colocar una semilla aleatoria

## Paso 5: Considerar los siguientes objetos: X_train, X_test, y_train, y_test

## Paso 6: Llamar a la librer√≠a / algoritmo
from sklearn.ensemble import RandomForestRegressor

## Paso 7: Entrenamos al algoritmo

### Paso 7.1: Definir un objeto (imputer) y los hiperpar√°metros del modelo
modelo_rf = RandomForestRegressor(n_estimators = 1000 , random_state = 123)

### Paso 7.2: Entrenamiento del algoritmo
modelo_rf.fit(X=X_train,y=y_train)


## Paso 8: Generamos las predicciones sobre los casos con data faltante

# El algoritmo entrenado => modelo_rf
# El algoritmo entrenado va generar predicciones sobre los datos que tienen info faltante (Appliances)
# El algoritmo entrenado necesita los features para generar las predicciones (data_1_df_imp_rf[var_[1:])

var_imp = modelo_rf.predict(data_1_df_imp_rf[var_[1:]])[data_1_df_imp_rf["Appliances"].isnull()]

## Paso 9: Imputamos con las predicciones del algoritmo
data_1_df_imp_rf.loc[data_1_df_imp_rf["Appliances"].isnull(),"Appliances"] = var_imp

## Paso 10: Verificamos los resultados despu√©s de la imputaci√≥n
print(f"Resumen de la variable despu√©s de ser imputada: \n{data_1_df_imp_rf[var_[0]].describe().transpose()}\n")
```
