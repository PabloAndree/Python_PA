```python
ruta_out = "/gdrive/My Drive/Colab Notebooks"
```

```python
data = pd.read_csv(ruta_in + "KAG_energydata_complete.csv", parse_dates=[0])
```

```python
data.head()
```

```python
data.columns.values
```

```python
var_ = ['Appliances', 'lights', 'T1', 'RH_1', 'T2', 'RH_2', 'T3',

       'RH_3', 'T4', 'RH_4', 'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7',

       'T8', 'RH_8', 'T9', 'RH_9', 'T_out', 'Press_mm_hg', 'RH_out',

       'Windspeed', 'Visibility', 'Tdewpoint']
```

```python
# Realizar un resumen con estad√≠ticas descriptivos de mis datos
resumen_data = data[var_].describe().T.reset_index()
resumen_data
```

```python
resumen_data['cv'] = np.round((resumen_data['std'] /np.abs(resumen_data['mean'])) * 100,2)
resumen_data
```

```python
resumen_data['as'] = np.round((3* (resumen_data['mean']- resumen_data['50%']))/resumen_data['std'],2)
resumen_data
```

```python
resumen_data.to_csv(ruta_out + "resumen_data_caso1.csv",index=False)



```

```python
x =data['Appliances'].values
x
```

```python
type(x)
```

```python
Q1 = np.percentile(x, 25)
Q3 = np.percentile(x, 75)
IQR = Q3 - Q1

LI = Q1 - 1.5 * IQR
LS = Q3 + 1.5 * IQR
```

```python
LI, LS
```

```python
data.shape
```

```python
data_1 = data[(data['Appliances']>= LI) & (data['Appliances']<= LS)]
data_1.shape
```

```python
px.histogram(data_1, x="Appliances", title = "Histogram - Appliances")
```

```python
px.box(data_1, y="Appliances", title = "Box plot - Appliances")
```
