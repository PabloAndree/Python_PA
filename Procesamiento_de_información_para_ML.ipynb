```python
#Imputacion por una medida de tendencia centrarl (promedio)
data_1_df_imp_mean = data_1_df.copy()
data_1_df_imp_mean['Appliances'] = data_1_df_imp_mean['Appliances'].fillna(data_1_df_imp_mean['Appliances'].mean())
```

```python
resumen_data_1_df_imp_mean = data_1_df_imp_mean[var_].describe().T.reset_index()
#pcf_df = porcentaje de data faltante a nivel de cada variable
resumen_data_1_df_imp_mean['pct_df'] = np.round((1-resumen_data_1_df_imp_mean['count']/data_1_df_imp_mean.shape[0])*100,2)
resumen_data_1_df_imp_mean
```

```python
#Imputacion por una medida de tendencia centrarl (mediana)
data_1_df_imp_median = data_1_df.copy()
data_1_df_imp_median['Appliances'] = data_1_df_imp_median['Appliances'].fillna(data_1_df_imp_median['Appliances'].median())
```

```python
resumen_data_1_df_imp_median = data_1_df_imp_median[var_].describe().T.reset_index()
#pcf_df = porcentaje de data faltante a nivel de cada variable
resumen_data_1_df_imp_median['pct_df'] = np.round((1-resumen_data_1_df_imp_median['count']/data_1_df_imp_median.shape[0])*100,2)
resumen_data_1_df_imp_median
```
