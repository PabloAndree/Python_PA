Simular los datos

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulamos los datos: Ingresos mensuales (media = 70, desviaci칩n est치ndar = 10)
np.random.seed(42)
ingresos = np.random.normal(70, 10, 1000)

# Visualizamos la distribuci칩n de ingresos
plt.hist(ingresos, bins=30, alpha=0.7, color="blue", density=True)
plt.title("Distribuci칩n de Ingresos Mensuales")
plt.xlabel("Ingreso (miles)")
plt.ylabel("Densidad de probabilidad")
plt.show()
```

Calculamos el intervalo

```python
# C치lculo de la media y desviaci칩n est치ndar muestral
media_muestral = np.mean(ingresos)
desviacion_muestral = np.std(ingresos, ddof=1)  # ddof=1 para la muestra
n = len(ingresos)

# Nivel de confianza del 95%
nivel_confianza = 0.95
z_critico = stats.norm.ppf(1 - (1 - nivel_confianza) / 2)

# Calcular el margen de error
margen_error = z_critico * (desviacion_muestral / np.sqrt(n))

# Calcular el intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

print(f"Intervalo de confianza del 95% para la media de ingresos: ({limite_inferior:.2f}, {limite_superior:.2f})")
```

Visualizar el intervalo de confianza

```python
# Visualizaci칩n del intervalo de confianza
plt.figure(figsize=(8,6))
plt.hist(ingresos, bins=30, alpha=0.7, color="blue", density=True)
plt.axvline(media_muestral, color="green", linestyle="--", label=f"Media muestral: {media_muestral:.2f}")
plt.axvline(limite_inferior, color="red", linestyle="--", label=f"L칤mite inferior: {limite_inferior:.2f}")
plt.axvline(limite_superior, color="red", linestyle="--", label=f"L칤mite superior: {limite_superior:.2f}")
plt.title("Intervalo de Confianza del 95% para los Ingresos")
plt.xlabel("Ingreso (miles)")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.show()
```
