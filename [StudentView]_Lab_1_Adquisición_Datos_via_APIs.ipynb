### **Paso 3:** Analizar datos de `n` personas

Dado que ya sabemos hace un request, ahora extramos data de `n` personas. Para entender el dataset aplique métodos para explorar los datos

1. Construye una función que extraiga del API `n` resultados.

```python
#TO DO
def get_dataNUsers(n):
  url = "https://randomuser.me/api/"
  lista_data = []
  for i in range(n):
    response = requests.get(url)
    data = response.json()
    lista_data.append(data)

  df = pd.json_normalize(lista_data, "results")
  return df
```

```python
def get_dataNUsers1(n):
  url = "https://randomuser.me/api/" + str(n)
  response = requests.get(url)
  data = response.json()
  df = pd.json_normalize(data, "results")
  return df
```

## **II. Adquisición de Datos a través de APIs con Autenticación**

Los enfoques de **autenticación** van desde lo más simple y directo, como los que usan claves API (API keys) o autenticación básica, hasta técnicas mucho más complejas y seguras, como `OAuth`.

Por lo general, llamar a una API sin credenciales o con las incorrectas devolverá un status code: `401 No autorizado` o `403 Prohibido`.

**API Keys**

Estas claves se utilizan para identificarse como usuario o cliente de la API y para rastrear su uso de la API. Las claves API generalmente se envían como un encabezado de solicitud (request header) o como un parámetro de consulta (query).

**API de la NASA:** una de las mejores colecciones de [API](https://apod.nasa.gov/apod/astropix.html) disponibles públicamente es la proporcionada por la NASA. Se puede obtener la imagen astronómica del día o imágenes tomadas por la Cámara de imágenes policromáticas de la Tierra (EPIC), entre otras.




### **Paso 1:** Definir la url

```python
#Definimos la url
url = "https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos"
```

### **Paso 2:** Construir los parámetros y request al servidor

```python
#Nasa otorga DEMO_KEY como key por default
api_key = "DEMO_KEY"
#Construir los párametros de consulta
query_params = {"api_key": api_key, "earth_date": "2023-10-03"}
#Hacer el request
response = requests.get(url, params=query_params)
response
```

### **Paso 3:** Convertir la respuesta a `json`

```python
#Recepcionar como dictionary
#TO DO
data = response.json()
data
```

```python
data.keys()
```

### **Paso 4:** Crear dataframe con los datos recolectados

```python
# Extraer los datos y guardarlos en un dataframe
df = pd.json_normalize(data, 'photos')
df.head()
```
