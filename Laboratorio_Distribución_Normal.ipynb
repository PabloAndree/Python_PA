**Uso de la tabla**

un conjunto de datos que representa las calificaciones de un examen de 100 estudiantes. Sabemos que las calificaciones siguen una distribución normal con una media de 70 y una desviación estándar de 10. Queremos saber qué porcentaje de estudiantes obtuvo una calificación mayor a 85.

x=85 (la calificación que nos interesa).

μ=70 (la media).

σ=10 (la desviación estándar).

```python
# Parámetros del problema
media = 70
desviacion_estandar = 10
calificacion = 85

# Calcular el valor Z
valor_z = (calificacion - media) / desviacion_estandar
print(f"El valor Z para una calificación de {calificacion} es: {valor_z:.2f}")
```

Con un valor Z de 1.5, podemos buscar este valor en la tabla de la distribución normal estándar. Las tablas de Z nos muestran la probabilidad acumulada desde el extremo izquierdo de la curva hasta el valor Z.



*   Z = 1.5 en la tabla da un valor de 0.9332. Esto significa que el 93.32% de los estudiantes tiene una calificación inferior a 85.
*   Para calcular el porcentaje de estudiantes que obtuvo una calificación mayor a 85, restamos este valor de 1:
                          P(X>85)=1−0.9332=0.0668
Esto significa que 6.68% de los estudiantes obtuvo una calificación mayor a 85.

Usar un conjunto de datos

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generar un conjunto de datos con una distribución normal (media 70, desviación estándar 10)
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=1000)
```

```python
# Visualización de la distribución de los datos
sns.histplot(data, kde=True)
plt.title('Distribución de Ingresos')
plt.xlabel('Ingreso (miles)')
plt.ylabel('Frecuencia')
plt.show()
```

Calcular el valor z

```python
# Cálculo de la media y la desviación estándar de los datos
media = np.mean(data)
desviacion_estandar = np.std(data)
```

```python
# Valor específico de ingreso
ingreso = 85
```

```python
# Cálculo del valor Z
valor_z = (ingreso - media) / desviacion_estandar
print(f"El valor Z para un ingreso de {ingreso} es: {valor_z:.2f}")
```

Estandarizar los datos

```python
# Estandarización de los datos
data_estandarizada = (data - media) / desviacion_estandar
```

```python
# Visualización de los datos estandarizados
sns.histplot(data_estandarizada, kde=True)
plt.title('Distribución de Ingresos Estandarizados')
plt.xlabel('Valor Z')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Comprobar que la media y desviación estándar de los datos estandarizados
print(f"Media de los datos estandarizados: {np.mean(data_estandarizada):.2f}")
print(f"Desviación estándar de los datos estandarizados: {np.std(data_estandarizada):.2f}")
```

Prueba de hipótesis (verificar normalidad)

```python
from scipy import stats
```

```python
# Realizar la prueba de Shapiro-Wilk
stat, p_value = stats.shapiro(data)
```

```python
# Mostrar resultados
print(f'Estadístico de Shapiro-Wilk: {stat:.4f}')
print(f'Valor p: {p_value:.4f}')
```

```python
# Verificar si los datos siguen una distribución normal (umbral de significancia 0.05)
if p_value > 0.05:
    print("No se puede rechazar la hipótesis nula. Los datos parecen seguir una distribución normal.")
else:
    print("Rechazamos la hipótesis nula. Los datos no siguen una distribución normal.")
```

Ejemplo del Teorema del Límite Central

```python
# Tomar muestras aleatorias del conjunto de datos
n_muestras = 1000  # número de muestras
tamano_muestra = 50  # tamaño de cada muestra
medias_muestrales = []
for _ in range(n_muestras):
    muestra = np.random.choice(data, size=tamano_muestra)
    medias_muestrales.append(np.mean(muestra))
```

```python
# Visualización de las medias muestrales
sns.histplot(medias_muestrales, kde=True)
plt.title('Distribución de Medias Muestrales')
plt.xlabel('Media de la muestra')
plt.ylabel('Frecuencia')
plt.show()
```
